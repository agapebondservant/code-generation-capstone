```python
#############################
# Manually upload LoRA model
#############################
checkpoint_path="./models/lora/ibm-granite_granite-4.0-h-tiny/final/checkpoint-1872"
model_id="ibm-granite/granite-4.0-h-tiny"
new_model_id="oaawofolu/granite-4.0-h-tiny_l"

model = AutoModelForCausalLM.from_pretrained(model_id, 
                                             
                                             trust_remote_code=True,
                                             
                                             device_map="cuda")

model = PeftModel.from_pretrained(model, checkpoint_path)

model = model.merge_and_unload()

tokenizer = AutoTokenizer.from_pretrained(model_id, local_files_only=True)
    
if tokenizer.pad_token is None:
    
    tokenizer.pad_token = tokenizer.eos_token

tokenizer.padding_side = 'right'

model.push_to_hub(new_model_id)

tokenizer.push_to_hub(new_model_id)
```
